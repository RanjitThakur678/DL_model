# ANN & TensorFlow 

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('bank_note_data.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
X = df[['Image.Var','Image.Skew','Image.Curt','Entropy']].values
y = df['Class'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense
```

```python
# model = Sequential([Dense(4,activation='relu'), Dense(2,activation='relu'),Dense(1)])
```

```python
model = Sequential()

model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='rmsprop',loss= 'mse')
```

```python
model.fit(x = X_train , y = y_train, epochs= 360)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_pred = model.predict(X_test)
```

```python
test_pred
```

```python
test_pred = pd.Series(test_pred.reshape(412,))
```

```python
test_pred
```

```python
pred_df = pd.DataFrame(y_test,columns=['Test True Y'])
```

```python
pred_df
```

```python
pred_df = pd.concat([pred_df,test_pred],axis=1)
```

```python
pred_df
```

```python
pred_df.columns = ['Test True Y ' , "Model prediction"]
```

```python
pred_df
```

```python
sns.boxplot(x = 'Test True Y ' , y = "Model prediction" , data = pred_df)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y '],pred_df["Model prediction"])
```
